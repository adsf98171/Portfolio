```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("df_total.xlsx")
```

```python
data = data[['Part_nbr', 'Lot_nbr', 'Layer', 'd_Layer','AVGX','AVGY','MacNo', 'MacRecipe', 'mtl_desc2', 'vend_nbr',
         'layer_2Thick', 'plate_mtl_desc1', 'plate_mtl_desc2','plate_vend_nbr', 'plate_layer_2Thick', 'plate_RC_value',
         'Residual Copper Rate', 'glass_fabric_thickness(mm)',
         'use_cu_upper_thickness', 'use_cu_lower_thickness', 'use_cu_type',
         'glass_fabric_type', 'short', 'long', 'd_PSVarX', 'd_PSVarY']]

data = data.copy()
data['layer_2Thick'] = data['layer_2Thick'].replace('None', np.nan).astype(float)
data['layer_2Thick'] = data['layer_2Thick'].astype('float')

data['plate_mtl_desc2'] = data['plate_mtl_desc2'].astype('float')

data['plate_layer_2Thick'] = pd.to_numeric(data['plate_layer_2Thick'], errors='coerce')
data = data.rename(columns={"glass_fabric_thickness(mm)": "glass_fabric_thickness_mm"})
data = data.rename(columns={"Residual Copper Rate": "Residual_Copper_Rate"})
df = data[data['d_PSVarX'] < 4000]
```

```python
def analyze_data_quality(df, group_col='Part_nbr_grouped', shrink_col='short', numeric_cols=None):
    if numeric_cols is None:
        numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        # 移除group_col和漲縮欄位避免重複分析
        numeric_cols = [c for c in numeric_cols if c not in [group_col, shrink_col]]

    # 1. 缺失率
    missing_pct = df.isna().mean()

    # 2. 各料號樣本數分布
    group_counts = df[group_col].value_counts()

    # 3. 漲縮異常判定 (每個group)
    def is_shrink_normal(group):
        max_val = group[shrink_col].max()
        min_val = group[shrink_col].min()
        mean_val = group[shrink_col].mean()
        # 判斷條件
        if (max_val - min_val) <= 3 and (0 <= mean_val <= 5):
            return True
        else:
            return False

    shrink_normal_flag = df.groupby(group_col).apply(is_shrink_normal)
    shrink_normal_flag = shrink_normal_flag.rename("shrink_normal")

    # 4. 數值欄位變異係數 CV = std / mean （避免 mean=0造成誤差）
    cv_dict = {}
    for col in numeric_cols:
        mean_val = df[col].mean()
        std_val = df[col].std()
        if mean_val != 0:
            cv = std_val / abs(mean_val)
        else:
            cv = np.nan
        cv_dict[col] = cv
    cv_series = pd.Series(cv_dict, name='CV')

    # 結果整理
    summary = {
        'missing_pct': missing_pct,
        'group_counts': group_counts,
        'shrink_normal_flag': shrink_normal_flag,
        'cv': cv_series
    }
    
    return summary
```

```python
# 範例用法
# df_filtered 是你清理過的DataFrame，Part_nbr_grouped是分組欄位，short是漲縮欄位
result = analyze_data_quality(df, group_col='Part_nbr', shrink_col='d_PSVarX', 
                              numeric_cols=['layer_2Thick', 'plate_mtl_desc2', 'plate_layer_2Thick', 'plate_RC_value', 
                                            'Residual_Copper_Rate', 'glass_fabric_thickness_mm', 'use_cu_upper_thickness', 
                                            'use_cu_lower_thickness', 'short'])

print("=== 缺失率 ===")
print(result['missing_pct'])

print("\n=== 各料號樣本數分布 (前20) ===")
print(result['group_counts'].head(10))

print("\n=== 漲縮異常判定 (True = 正常) ===")
print(result['shrink_normal_flag'].value_counts())

print("\n=== 數值欄位變異係數 CV ===")
print(result['cv'])
```
