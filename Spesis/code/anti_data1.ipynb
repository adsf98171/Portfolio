```python
import pandas as pd
import numpy as np
import re
```

```python
df2024 = pd.read_csv(r'C:\Users\482525\Desktop\敗血症資料\2024\2024Sepsis00101.csv', encoding='big5', dtype={'INTIME': str, 'EMGOUTTIME': str, 0: str})
df2025 = pd.read_csv(r'C:\Users\482525\Desktop\敗血症資料\2025\2025Sepsis00101.csv', encoding='big5', dtype={'INTIME': str, 'EMGOUTTIME': str, 0: str})

df2024['INTIME'] = pd.to_datetime(df2024['INTIME'], format='%Y%m%d%H%M%S', errors='coerce')
df2024['EMGOUTTIME'] = pd.to_datetime(df2024['EMGOUTTIME'], format='%Y%m%d%H%M%S', errors='coerce')
df2025['INTIME'] = pd.to_datetime(df2025['INTIME'], format='%Y%m%d%H%M%S', errors='coerce')
df2025['EMGOUTTIME'] = pd.to_datetime(df2025['EMGOUTTIME'], format='%Y%m%d%H%M%S', errors='coerce')

table1 = pd.concat([df2024, df2025], ignore_index=True)
```

```python
table1 = table1.dropna(how='all')
table1 = table1[table1['ACCOUNTNO'].notna()]
table1 = table1.loc[:, ~table1.columns.str.contains('^Unnamed')]
```
